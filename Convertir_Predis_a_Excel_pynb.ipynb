```python
!pip install camelot-py[cv] ghostscript openpyxl

import camelot
import pandas as pd
from google.colab import files

# 1. Subir TODOS los PDFs de 2006 a 2020
uploaded = files.upload()
print("Archivos subidos:", list(uploaded.keys()))

# 2. Procesar cada archivo y consolidar
df_total = []

for nombre_pdf in uploaded.keys():
    try:
        # Detectar el año a partir del nombre del archivo
        # Ej: "REGISTROS PREDIS 2006.pdf" → 2006
        anio = "".join([c for c in nombre_pdf if c.isdigit()])

        # Extraer tablas con Camelot
        tablas = camelot.read_pdf(nombre_pdf, pages="all", flavor="stream")
        df_list = [t.df for t in tablas]
        df = pd.concat(df_list, ignore_index=True)

        # Agregar columna Año
        df["Año"] = anio

        # Guardar en lista para consolidar
        df_total.append(df)

        # Guardar Excel individual por año
        df.to_excel(f"registros_predis_{anio}.xlsx", index=False)
        print(f"✅ Procesado {nombre_pdf}, {len(df)} filas")
    except Exception as e:
        print(f"⚠️ Error con {nombre_pdf}: {e}")

# 3. Consolidar todos los años en un solo archivo
df_final = pd.concat(df_total, ignore_index=True)
df_final.to_excel("registros_predis_2006_2020.xlsx", index=False)

# 4. Descargar el consolidado
files.download("registros_predis_2006_2020.xlsx")
```
